```python
#define decrypt function
def decrypt(ciphertext):  
    #assign plaintext variable
    plaintext = ""
    #start index at character
    i = 0
    
    #loop will run until the end of the ciphertext length
    while i < len(ciphertext):
        
        for i in range(len(ciphertext)):
            #if index position in ciphertext is even
            if i % 2 == 0: 
                #change ciphertext to ascii
                cipher_ascii = ord(ciphertext[i])
                #change cipher ascii to plain ascii from info provided in the encrypt flowchart
                plain_ascii = cipher_ascii - 1
            
            #otherwise, if index position in ciphertext string is odd
            else:
                #change ciphertext to ascii
                cipher_ascii = ord(ciphertext[i])
                #change cipher ascii to plain ascii from info provided in the encrpyt flowchart
                plain_ascii = cipher_ascii + 1
                
            #translate plain ascii to corresponding character
            plaintext += chr(plain_ascii)
            #iteration for loop
            i += 1
            
        #return decrypted message
        return plaintext
```

```python
import random

#function for our game
def hangmanGame():
    #number of lives user starts with
    lives_left = 10 

    #list of letters the user has guessed
    already_guessed = []
    
    #list of already encoded hangman words
    words = ["WHDSPQZ", "XHO", "TTDBFRT", "QQJYF", "ENQD", "DNPK", "CNTR", "EHWHOD", "TSVMOHOF", "XNOCFQGTM"]

    #variable that randomly chooses a word from the list above to use for the game
    hangman_word = random.choice(words)
    
    #assinging the randomly chosen hangman word to the ciphertext variable to call later for the decryption function
    ciphertext = hangman_word
    
    #creates the secret word as a list, with dashes equal to the secret word length
    secret_hangman_word = [' -' for i in range(len(hangman_word))]
    
    while lives_left > 0:
        
        #print line #1 updates user on number of lives left and guesses so far
        print(f"You have {lives_left} lives left and you have used these letters:", *already_guessed)
        
        #print line #2 prints dashes for unguessed letters from the hidden word
        print("Current word: ", *secret_hangman_word)
        
        #print line #3 prompts user to guess lettesr
        user_guess = input("Guess a letter: ")
        user_guess = user_guess.upper() #turns user guess into uppercase to check against the case of our word list
       
        #if user makes a repeat guess, it doesn't count against their lives
        if user_guess in already_guessed:
                print(f"You already guessed {user_guess} \n")
                continue
                
        #if user makes a correct guess for the first time
        if user_guess in hangman_word:
            
            print(f"Nice choice, {user_guess} is in the word. \n")
                
            #add letter to guessing list
            already_guessed.append(user_guess)
            
            #replace list of dashes (by index) with guessed letter
            for i in range(len(hangman_word)):
                if hangman_word[i] == user_guess:
                    secret_hangman_word[i] = user_guess
                    
            #to determine a win: joins the word the user is guessing together so we can compare it to the secret word.
            joined_hangman_word = ''.join(secret_hangman_word)
            
             #if the joined word the user is guessing is the same as the secret word, they win
            if joined_hangman_word == hangman_word:
                
                #call decrypt function
                decrypt(hangman_word)
                
                #print player win message
                print(f'''Congratulations, you guessed the word: {hangman_word}
                This was the encoded message!
                The decoded message says: {decrypt(ciphertext)}!''')
                
                #making lives left = -1 will end our game and restart the loop, asking player if they want to play again
                lives_left = -1
                break
                
        #if user guesses incorrectly 
        else:
            
            print(f"{user_guess} is not in the word. \n")
            #add letter to guessing list
            already_guessed.append(user_guess)
            #remove one life, womp womp
            lives_left -= 1
                      
    #when user runs out of lives, the game ends
    if lives_left == 0:
        
        #print player lose message 
        print(f'''Womp womp. You ran out of lives and lost the game. 
The word was {hangman_word}. You'll have to guess the word correctly to decode the message. ''')
  
```

```python

#input user name to return basic greeting
user_name = input("What is your name? ")

#print brief game introduction for player
print(f'''\nHello, {user_name}, I want to play a game!
You have 10 lives to guess the randomly selected word.
If you win, I will decode the message. \n ''')
    
#this loop will keep running to check and see if the user (still) wants to play after the games end
while True:

    #input asks user if they want to play the game
    play_start = input('''Ready to play?
    1. Yes
    2. No
    ''')
    
#if user wants to play
    if play_start == '1':
        
        print("Great, let's start: \n")

        #Call function to start the game
        hangmanGame()

    #if user doesn't want to play
    elif play_start == '2':
        print("Okay, maybe next time. Bye!")
        break
    
    #if user chooses invalid input, program will print invalid statement and loop again
    else:
        print("Sorry, that is not a valid input, please choose again. ")
```
